# Retrieval Accuracy Benchmark

Runs `AccuracyEvaluator` across all 6 agent variants and saves results.

| # | File | Model | Components |
|---|------|-------|-----------|
| 1 | `llama_api` | llama3.2 | full guide |
| 2 | `llama_api_semantic_retriever` | llama3.2 | semantic |
| 3 | `llama_api_final` | llama3.2 | semantic + date parser + checks |
| 4 | `gpt_api` | gpt-4o-mini | full guide |
| 5 | `gpt_api_semantic_retriever` | gpt-4o-mini | semantic |
| 6 | `gpt_api_final` | gpt-4o-mini | semantic + checks |

## 0. Setup

```python
import os, time
from retrieval_accuracy_test import AccuracyEvaluator

# agent classes
from llama_api                  import FredLLMAgent
from llama_api_semantic_retriever import FredLLMAgent  as FredLLMAgent_semantic
from llama_api_final            import FredLLMAgent    as FredLLMAgent_final
from gpt_api                    import OpenAIFredAgent
from gpt_api_semantic_retriever import OpenAIFredAgent as OpenAIFredAgent_semantic
from gpt_api_final              import OpenAIFredAgent as OpenAIFredAgent_final

TEST_FILE   = "data/QA_test.json"
LLAMA_DIR   = "files/llama3.2"
GPT_DIR     = "files/gpt-4o-mini"

os.makedirs(LLAMA_DIR, exist_ok=True)
os.makedirs(GPT_DIR,   exist_ok=True)

def make_evaluator(agent) -> AccuracyEvaluator:
    """Inject a pre-built agent into a fresh AccuracyEvaluator."""
    # Use llama3.2 as a dummy model arg — agent is overwritten immediately
    ev = AccuracyEvaluator.__new__(AccuracyEvaluator)
    ev.agent      = agent
    ev.test_cases = []
    ev.load_test_cases(TEST_FILE)
    return ev

def run_and_save(evaluator: AccuracyEvaluator, out_path: str):
    t0 = time.time()
    results = evaluator.run_all_tests()
    print(f"Execution time: {time.time() - t0:.2f}s")
    evaluator.export_results(results, out_path)
    return results
```

```python
res1, res2, res3, res4, res5, res6 = {}, {}, {}, {}, {}, {}
```

---
## 1. llama_api  (full guide)

```python
ev1     = make_evaluator(FredLLMAgent(model="llama3.2", verbose=False))
res1    = run_and_save(ev1, f"{LLAMA_DIR}/retrieval_eval/llama_api.json")
```

---
## 2. llama_api_semantic_retriever

```python
ev2     = make_evaluator(FredLLMAgent_semantic(model="llama3.2", verbose=False))
res2    = run_and_save(ev2, f"{LLAMA_DIR}/retrieval_eval/llama_api_semantic_retriever.json")
```

---
## 3. llama_api_final  (semantic + self-checks + date parser)

```python
ev3     = make_evaluator(FredLLMAgent_final(model="llama3.2", verbose=False))
res3    = run_and_save(ev3, f"{LLAMA_DIR}/retrieval_eval/llama_api_final.json")
```

---
## 4. gpt_api  (full guide)

```python
ev4     = make_evaluator(OpenAIFredAgent(verbose=False))
res4    = run_and_save(ev4, f"{GPT_DIR}/retrieval_eval/gpt_api.json")
```

---
## 5. gpt_api_semantic_retriever

```python
ev5     = make_evaluator(OpenAIFredAgent_semantic(verbose=False))
res5    = run_and_save(ev5, f"{GPT_DIR}/retrieval_eval/gpt_api_semantic_retriever.json")
```

---
## 6. gpt_api_final  (semantic + self-checks + date parser)

```python
ev6     = make_evaluator(OpenAIFredAgent_final(verbose=False))
res6    = run_and_save(ev6, f"{GPT_DIR}/retrieval_eval/gpt_api_final.json")
```

## 7. Summary comparison

```python
import json

if not res1:
    with open("files/llama3.2/retrieval_eval/llama_api.json") as f:
        res1 = json.load(f)
if not res2:
    with open("files/llama3.2/retrieval_eval/llama_api_semantic_retriever.json") as f:
        res2 = json.load(f)
if not res3:
    with open("files/llama3.2/retrieval_eval/llama_api_final.json") as f:
        res3 = json.load(f)
if not res4:
    with open("files/gpt-4o-mini/retrieval_eval/gpt_api.json") as f:
        res4 = json.load(f)
if not res5:
    with open("files/gpt-4o-mini/retrieval_eval/gpt_api_semantic_retriever.json") as f:
        res5 = json.load(f)
if not res6:
    with open("files/gpt-4o-mini/retrieval_eval/gpt_api_final.json") as f:
        res6 = json.load(f)
```

```python
import pandas as pd

rows = [
    ("llama_api",                    res1),
    ("llama_api_semantic_retriever", res2),
    ("llama_api_final",              res3),
    ("gpt_api",                      res4),
    ("gpt_api_semantic_retriever",   res5),
    ("gpt_api_final",                res6),
]

records = []
for name, res in rows:
    s = res["summary"]
    records.append({
        "agent":              name,
        "total":              s["total_tests"],
        "passed":             s["successful_tests"],
        "failed":             s["failed_tests"],
        "series_id_score":    s["avg_series_id_score"],
        "date_range_score":   s["avg_date_range_score"],
        "overall_score":      s["avg_overall_score"],
    })

df = pd.DataFrame(records).set_index("agent")
df
```
