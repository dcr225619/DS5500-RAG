```python
import os, time
import re
from retrieval_accuracy_test import AccuracyEvaluator

# agent classes
from llama_api                  import FredLLMAgent
from llama_api_semantic_retriever import FredLLMAgent  as FredLLMAgent_semantic
from llama_api_final            import FredLLMAgent    as FredLLMAgent_final
from gpt_api                    import OpenAIFredAgent
from gpt_api_semantic_retriever import OpenAIFredAgent as OpenAIFredAgent_semantic
from gpt_api_final              import OpenAIFredAgent as OpenAIFredAgent_final

base_dir = os.getcwd()

TEST_FILE   = os.path.join(base_dir, "../data/QA_test.json")
LLAMA_DIR   = os.path.join(base_dir, "../files/llama3.2")
GPT_DIR     = os.path.join(base_dir, "../files/gpt-4o-mini")

os.makedirs(LLAMA_DIR, exist_ok=True)
os.makedirs(GPT_DIR,   exist_ok=True)

def make_evaluator(agent) -> AccuracyEvaluator:
    """Inject a pre-built agent into a fresh AccuracyEvaluator."""
    # Use llama3.2 as a dummy model arg — agent is overwritten immediately
    ev = AccuracyEvaluator.__new__(AccuracyEvaluator)
    ev.agent      = agent
    ev.test_cases = []
    ev.load_test_cases(TEST_FILE)
    return ev

def run_and_save(evaluator: AccuracyEvaluator, out_path: str):
    t0 = time.time()
    results = evaluator.run_all_tests()
    print(f"Execution time: {time.time() - t0:.2f}s")
    evaluator.export_results(results, out_path)
    return results
```

## Checks only and Date Parser only

```python
import json
from datetime import datetime, timedelta


def load_json(path):
    with open(path, "r", encoding="utf-8") as f:
        return json.load(f)


def resolve_date(date_spec, is_end=False):
    """Resolve relative date strings (e.g. '1y', '6m', 'today') to YYYY-MM-DD."""
    if not date_spec:
        return ""
    if date_spec == "today":
        return datetime.today().strftime("%Y-%m-%d")
    if date_spec.endswith("y"):
        return (datetime.today() - timedelta(days=365 * int(date_spec[:-1]))).strftime("%Y-%m-%d")
    if date_spec.endswith("m"):
        return (datetime.today() - timedelta(days=30 * int(date_spec[:-1]))).strftime("%Y-%m-%d")
    return date_spec


def compare_dates(date1_str, date2_str, tolerance_days=30):
    """
    Score how well date1 <= date2 (date1 expected earlier).
    Returns 1.0 for exact/within tolerance, 0.0 if date2 < date1.
    """
    if not date1_str or not date2_str:
        return 0.0
    try:
        d1 = datetime.strptime(date1_str, "%Y-%m-%d")
        d2 = datetime.strptime(date2_str, "%Y-%m-%d")
        diff = (d2 - d1).days
        if diff >= 0:
            return max(1.0 - max(diff - tolerance_days, 0) / 100, 0.0)
        else:
            return 0.0
    except ValueError:
        return 0.0


def evaluate_series_id(actual_calls, expected_ids):
    actual_set = {c["series_id"] for c in actual_calls}
    expected_set = set(expected_ids)

    correct = actual_set & expected_set
    precision = len(correct) / len(actual_set) if actual_set else 0.0
    recall = len(correct) / len(expected_set) if expected_set else 1.0
    f1 = (2 * precision * recall / (precision + recall)) if (precision + recall) > 0 else 0.0

    return {
        "score": round(f1, 4),
        "correct": sorted(correct),
        "missing": sorted(expected_set - actual_set),
        "extra": sorted(actual_set - expected_set),
    }


def evaluate_date_range(actual_calls, expected_date_range, tolerance_days=30):
    if not expected_date_range or not actual_calls:
        return {"score": 1.0, "details": []}

    exp_start = resolve_date(
        expected_date_range.get("relative_start") or expected_date_range.get("start", "")
    )
    exp_end = resolve_date(
        expected_date_range.get("relative_end") or expected_date_range.get("end", ""),
        is_end=True,
    )

    scores, details = [], []
    for call in actual_calls:
        act_start = call.get("start_date", "")
        act_end = call.get("end_date", "")

        # start: actual_start should be <= expected_start (covers early enough)
        start_score = compare_dates(act_start, exp_start, tolerance_days)
        # end: expected_end should be <= actual_end (covers late enough)
        end_score = compare_dates(exp_end, act_end, tolerance_days)

        call_score = (start_score + end_score) / 2
        scores.append(call_score)
        details.append({
            "series_id": call.get("series_id"),
            "actual_start": act_start,
            "expected_start": exp_start,
            "start_score": round(start_score, 4),
            "actual_end": act_end,
            "expected_end": exp_end,
            "end_score": round(end_score, 4),
            "call_score": round(call_score, 4),
        })

    return {
        "score": round(sum(scores) / len(scores), 4),
        "details": details,
    }


def evaluate_case(pred, ground_truth, tolerance_days=30):
    """
    Evaluate one prediction record against its ground truth entry.
    Returns a result dict with series_id and date_range evaluations.
    """
    qid = ground_truth.get("question_id", pred.get("question", "unknown"))
    tool_call_required = ground_truth.get("tool_call_required", True)
    actual_calls = pred.get("tool_calls", [])

    # ── Non-tool-call questions ──────────────────────────────────────────────
    if not tool_call_required:
        made_call = bool(actual_calls)
        return {
            "question_id": qid,
            "question": ground_truth.get("question", ""),
            "tool_call_required": False,
            "success": not made_call,
            "error": "Unexpected tool call made" if made_call else None,
            "series_id_score": 0.0 if made_call else 1.0,
            "date_range_score": 0.0 if made_call else 1.0,
            "overall_score": 0.0 if made_call else 1.0,
        }

    # ── Tool-call questions ──────────────────────────────────────────────────
    if not pred.get("success") or not actual_calls:
        return {
            "question_id": qid,
            "question": ground_truth.get("question", ""),
            "tool_call_required": True,
            "success": False,
            "error": "No tool call extracted",
            "series_id_score": 0.0,
            "date_range_score": 0.0,
            "overall_score": 0.0,
        }

    expected_ids = ground_truth.get("expected_series_ids", [])
    expected_dr = ground_truth.get("expected_date_range", {})

    series_eval = evaluate_series_id(actual_calls, expected_ids)
    date_eval = evaluate_date_range(actual_calls, expected_dr, tolerance_days)
    overall = round(series_eval["score"] * 0.5 + date_eval["score"] * 0.5, 4)

    return {
        "question_id": qid,
        "question": ground_truth.get("question", ""),
        "tool_call_required": True,
        "success": True,
        "actual_calls": actual_calls,
        "series_id_evaluation": series_eval,
        "date_range_evaluation": date_eval,
        "series_id_score": series_eval["score"],
        "date_range_score": date_eval["score"],
        "overall_score": overall,
    }


def run_evaluation(
    pred_path,
    gt_path,
    output_path=None,
    tolerance_days=30,
):
    """
    Args:
        pred_path:   path to QA_test_llama_api_date_only.json (model outputs)
        gt_path:     path to QA_test.json (ground truth with expected_series_ids etc.)
        output_path: optional path to write the result JSON
        tolerance_days: date tolerance window
    """
    preds = load_json(pred_path)
    gts = load_json(gt_path)

    # Build lookup by question text (fallback) or question_id
    gt_by_id = {g.get("question_id"): g for g in gts if "question_id" in g}
    gt_by_q = {g["question"]: g for g in gts}

    results = []
    for pred in preds:
        qid = pred.get("question_id")
        question = pred.get("question", "")
        gt = gt_by_id.get(qid) or gt_by_q.get(question)

        if gt is None:
            results.append({
                "question_id": qid or question,
                "question": question,
                "success": False,
                "error": "No matching ground truth found",
                "overall_score": 0.0,
            })
            continue

        results.append(evaluate_case(pred, gt, tolerance_days))

    # ── Summary ──────────────────────────────────────────────────────────────
    successful = [r for r in results if r["success"]]
    tool_cases = [r for r in results if r.get("tool_call_required", True)]
    non_tool_cases = [r for r in results if not r.get("tool_call_required", True)]

    def avg(lst, key):
        vals = [x[key] for x in lst if key in x]
        return round(sum(vals) / len(vals), 4) if vals else 0.0

    summary = {
        "total": len(results),
        "successful": len(successful),
        "failed": len(results) - len(successful),
        "tool_call_tests": len(tool_cases),
        "tool_call_passed": sum(1 for r in tool_cases if r["success"]),
        "non_tool_call_tests": len(non_tool_cases),
        "non_tool_call_passed": sum(1 for r in non_tool_cases if r["success"]),
        "avg_series_id_score": avg(successful, "series_id_score"),
        "avg_date_range_score": avg(successful, "date_range_score"),
        "avg_overall_score": avg(successful, "overall_score"),
    }

    # Print summary
    sep = "=" * 55
    print(f"\n{sep}")
    print("EVALUATION SUMMARY")
    print(sep)
    print(f"Total tests:           {summary['total']}")
    print(f"Successful:            {summary['successful']}")
    print(f"Failed:                {summary['failed']}")
    print(f"\nBreakdown:")
    print(f"  Tool-call tests:     {summary['tool_call_tests']} "
          f"(passed: {summary['tool_call_passed']})")
    print(f"  Non-tool-call tests: {summary['non_tool_call_tests']} "
          f"(passed: {summary['non_tool_call_passed']})")
    print(f"\nAverage Scores (successful tests only):")
    print(f"  Series ID score:     {summary['avg_series_id_score']}")
    print(f"  Date range score:    {summary['avg_date_range_score']}")
    print(f"  Overall score:       {summary['avg_overall_score']}")
    print(sep)

    output = {"summary": summary, "results": results}

    if output_path:
        with open(output_path, "w", encoding="utf-8") as f:
            json.dump(output, f, indent=2, ensure_ascii=False)
        print(f"\nResults saved to: {output_path}")

    return output
```
